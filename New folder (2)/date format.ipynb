```python
import pandas as pd
import datetime

# Load the data into a pandas DataFrame
df = pd.read_csv("/content/003 TachyHealth-ClaimsDetails - 06022023.csv", encoding='ISO-8859-1')

def convert_date(date_str):
    day, month, year = map(int, date_str.split("/"))
    if year >= 0 and year <= 23:
        year += 2000
    else:
        year += 1900
    return pd.to_datetime(f"{day}/{month}/{year}", format='%d/%m/%Y')


df['DateofBirth'] = df['DateofBirth'].apply(convert_date)

# Convert the datetime format back to a string in the desired format
df['DateofBirth'] = df['DateofBirth'].dt.strftime('%d/%m/%Y')
df['DateofBirth'] = pd.to_datetime(df['DateofBirth'], format='%d/%m/%Y')
df[['DateofBirth']].to_csv("output.csv", index=False)
```
