```python
import pandas as pd

df = pd.read_excel('/content/SFDA complete list 26 Dec 2022.xlsx', sheet_name='Sheet3')
```

```python
df
```

```python
# Convert all values to strings 
df['concat'] = df.apply(lambda row: '/'.join(row.dropna().astype(str).values), axis=1)

# For completely blank rows, fill in a single space
df.loc[df['concat'].isna(), 'concat'] = ' ' 

print(df)
```

```python
df.to_excel('output.xlsx')
```
