```python
import pandas as pd
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("/content/endoscopic vs histopathology.csv")
df
```

```python
# Map labels to binary values
df['histo_binary'] = df['histopathology'].map({'atrophic (spotty)': 1, 'no atrophy': 0})
df['endo_binary'] = df['Endoscopic score'].map({'atrophic (spotty)': 1, 'no atrophy': 0})

# Check if positive samples exist for histopathology
if 1 in df['histo_binary'].values:
    # Calculate and plot histopathology ROC curve
    y_true_histo = df['histo_binary']
    predictor_histo = df['Endoscopic score']
    fpr_histo, tpr_histo, thresholds_histo = roc_curve(y_true_histo, predictor_histo, pos_label=1)
    auc_histo = auc(fpr_histo, tpr_histo)

    plt.plot(fpr_histo, tpr_histo, color='darkorange', lw=2, label='Histo ROC (AUC = %0.2f)' % auc_histo)
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.title('Histopathology ROC Curve')
    plt.legend()
    plt.show()
else:
    print("No positive samples found for histopathology. ROC curve cannot be plotted.")

# Check if positive samples exist for endoscopy
if 1 in df['endo_binary'].values:
    # Calculate and plot endoscopy ROC curve
    y_true_endo = df['endo_binary']
    predictor_endo = df['histopathology']
    fpr_endo, tpr_endo, thresholds_endo = roc_curve(y_true_endo, predictor_endo, pos_label=1)
    auc_endo = auc(fpr_endo, tpr_endo)

    plt.plot(fpr_endo, tpr_endo, color='darkorange', lw=2, label='Endo ROC (AUC = %0.2f)' % auc_endo)
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.title('Endoscopy ROC Curve')
    plt.legend()
    plt.show()
else:
    print("No positive samples found for endoscopy. ROC curve cannot be plotted.")
```
