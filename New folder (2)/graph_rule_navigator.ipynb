```python
!pip install networkx
# Import libraries
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Read Excel file
df = pd.read_excel('Acute embolism and thrombosis of inferior vena cava.xlsx')

# Create graph
G = nx.DiGraph()

# Add nodes
node_features = df.columns
G.add_nodes_from(node_features)

# Add edges
for i in range(len(node_features)-1):
  G.add_edge(node_features[i], node_features[i+1])

# Visualize
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your data is in a DataFrame 'df'
# For this example, let's create a mock DataFrame
data = {
    'Template Name': ['Acute embolism and thrombosis of inferior vena cava', 'Acute myocardial infarction'],
    'Procedure Code': ['71250', '78414']
}
df = pd.DataFrame(data)

# Create a graph
G = nx.Graph()

# Add nodes and edges from your data
for index, row in df.iterrows():
    G.add_node(row['Template Name'], color='red')
    G.add_node(row['Procedure Code'], color='blue')
    G.add_edge(row['Template Name'], row['Procedure Code'])

# Draw the graph
colors = [node[1]['color'] for node in G.nodes(data=True)]
nx.draw(G, with_labels=True, node_color=colors)
plt.show()
```
