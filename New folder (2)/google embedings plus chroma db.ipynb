```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
import gc
from tqdm.notebook import tqdm
import time

# Define the batch size
batch_size = 1000

# Load the dataframes if they are not already loaded
df1 = pd.read_pickle('/content/drive/MyDrive/achi_embedings.pkl')
df2 = pd.read_pickle('/content/drive/MyDrive/df_2.pkl')

# Initialize lists to store the similarity scores and texts
text_df1 = []
text_df2 = []
similarities = []

# Initialize the start time
start_time = time.time()

# Iterate over each batch of embeddings in df1
for i in tqdm(range(0, len(df1), batch_size), total=((len(df1) - 1) // batch_size + 1), desc='Processing batches'):

    # Record the time at the start of each iteration
    batch_start_time = time.time()

    # Load batch1 into memory
    batch1 = df1[i:i+batch_size]

    # Calculate the cosine similarity for each pair of embeddings in the batch
    similarity_matrix = cosine_similarity(batch1['Embeddings'].tolist(), df2['Embeddings'].tolist())

    # Iterate over each similarity score in the similarity matrix
    for k in range(similarity_matrix.shape[0]):
        for l in range(similarity_matrix.shape[1]):
            # Append the similarity score and texts to the lists
            text_df1.append(batch1.iloc[k]['Text'])
            text_df2.append(df2.iloc[l]['Text'])
            similarities.append(similarity_matrix[k][l])

    # Clear batch1 from memory
    del batch1
    gc.collect()

    # Calculate and print the elapsed time for this iteration
    batch_elapsed_time = time.time() - batch_start_time
    print(f"Time elapsed for batch: {batch_elapsed_time:.2f} seconds")

    # Calculate and print the estimated remaining time
    completed_iterations = i // batch_size + 1
    total_iterations = (len(df1) - 1) // batch_size + 1
    average_time_per_iteration = (time.time() - start_time) / completed_iterations
    estimated_remaining_time = average_time_per_iteration * (total_iterations - completed_iterations)
    print(f"Estimated remaining time: {estimated_remaining_time:.2f} seconds")

# Create the DataFrame from the lists
similarity_scores = pd.DataFrame({
    'Text_df1': text_df1,
    'Text_df2': text_df2,
    'Similarity': similarities
})

print(similarity_scores)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
import gc
from tqdm.notebook import tqdm
import time

# Define the batch size
batch_size = 1000

# Load the dataframes if they are not already loaded
df1 = pd.read_pickle('/content/drive/MyDrive/achi_embedings.pkl')
df2 = pd.read_pickle('/content/drive/MyDrive/df_2.pkl')
```

```python
df1 = df1.dropna()
```

```python
df2 = df2.dropna()
```

```python
# Initialize lists to store the maximum similarity scores and texts
text_df1 = []
text_df2 = []
max_similarities = []

# Initialize the start time
start_time = time.time()

# Iterate over each batch of embeddings in df1
for i in tqdm(range(0, len(df1), batch_size), total=((len(df1) - 1) // batch_size + 1), desc='Processing batches'):

    # Record the time at the start of each iteration
    batch_start_time = time.time()

    # Load batch1 into memory
    batch1 = df1[i:i+batch_size]

    # Iterate over each row in the batch
    for j in range(len(batch1)):

        # Record the time at the start of each row
        row_start_time = time.time()

        # Calculate the cosine similarity for this row against all of df2
        similarity_row = cosine_similarity([batch1.iloc[j]['Embeddings']], df2['Embeddings'].tolist())

        # Find the index of the maximum similarity score
        max_index = np.argmax(similarity_row)

        # Append the maximum similarity score and corresponding texts to the lists
        text_df1.append(batch1.iloc[j]['Text'])
        text_df2.append(df2.iloc[max_index]['Text'])
        max_similarities.append(similarity_row[0][max_index])

        # Calculate and print the elapsed time for this row
        row_elapsed_time = time.time() - row_start_time
        print(f"Time elapsed for row: {row_elapsed_time:.2f} seconds")

    # Clear batch1 from memory
    del batch1
    gc.collect()

    # Calculate and print the elapsed time for this batch
    batch_elapsed_time = time.time() - batch_start_time
    print(f"Time elapsed for batch: {batch_elapsed_time:.2f} seconds")

    # Calculate and print the estimated remaining time for the rest of the batches
    completed_batches = i // batch_size + 1
    total_batches = (len(df1) - 1) // batch_size + 1
    average_time_per_batch = (time.time() - start_time) / completed_batches
    estimated_remaining_time_batches = average_time_per_batch * (total_batches - completed_batches)
    print(f"Estimated remaining time for the rest of the batches: {estimated_remaining_time_batches:.2f} seconds")

# Create the DataFrame from the lists
similarity_scores = pd.DataFrame({
    'Text_df1': text_df1,
    'Text_df2': text_df2,
    'Maximum Similarity': max_similarities
})

print(similarity_scores)
```
