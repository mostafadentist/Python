```python
import pandas as pd
from tqdm import tqdm
import numpy as np

# Function to standardize date format
def standardize_date(val):
    # Define possible date formats
    date_formats = (
        '%Y-%m-%d %H:%M:%S',
        '%d-%m-%Y %H:%M',
        '%Y-%m-%d',
        '%d-%m-%Y',
        '%m-%d-%Y %H:%M:%S',
        '%Y-%d-%m %H:%M:%S',
        '%m-%d-%Y',
        '%Y-%d-%m'
    )
    # Try different date formats
    for fmt in date_formats:
        try:
            return pd.to_datetime(val, format=fmt).strftime('%d/%m/%Y')
        except (ValueError, TypeError):
            continue

    # If all else fails, print a warning and return NaT
    print(f"Warning: unrecognized date {val}")
    return pd.NaT

# Load the data
df = pd.read_excel('/content/Dates issue of Tawnyina.xlsx')

# Identify the columns that contain dates
date_columns = df.select_dtypes(include=[np.datetime64]).columns.tolist()

# Standardize the dates
for column in tqdm(date_columns):
    df[column] = df[column].apply(standardize_date)

# Save the cleaned data to a new Excel file
df.to_excel('/content/cleaned_dates_tawnyina.xlsx', index=False)
```
