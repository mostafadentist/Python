```python
# Load the documents
# link to download data : https://www.kaggle.com/datasets/ashishraut64/indian-startups-top-300?resource=download
loader = CSVLoader(file_path='/content/1.csv')
```

```python
df = pd.read_csv('/content/1.csv')
df
```

```python
# Create an index using the loaded documents
index_creator = VectorstoreIndexCreator()
docsearch = index_creator.from_loaders([loader])
```

```python
# Create a question-answering chain using the index
chain = RetrievalQA.from_chain_type(llm=OpenAI(), chain_type="stuff", retriever=docsearch.vectorstore.as_retriever(), input_key="question")
```

```python
# Pass a query to the data
query = "which PhysicianID has the highest number_inpatient_benefit_code ?"
response = chain({"question": query})
print(response['result'])
```
