```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
import gc
from tqdm.notebook import tqdm
import time

# Define the batch size
batch_size = 1000

# Load the dataframes if they are not already loaded
df1 = pd.read_pickle('/content/drive/MyDrive/achi_embedings.pkl')
df2 = pd.read_pickle('/content/drive/MyDrive/df_2.pkl')

# Initialize lists to store the similarity scores and texts
text_df1 = []
text_df2 = []
similarities = []

# Initialize the start time
start_time = time.time()

# Iterate over first three rows of df1
for i in tqdm(range(3), total=3, desc='Processing batches'):

    # Record the time at the start of each iteration
    batch_start_time = time.time()

    # Load batch1 into memory
    batch1 = df1[i:i+1]

    # Iterate over each row in the batch
    for j in range(len(batch1)):

        # Record the time at the start of each row
        row_start_time = time.time()

        # Calculate the cosine similarity for this row against all of df2
        similarity_row = cosine_similarity([batch1.iloc[j]['Embeddings']], df2['Embeddings'].tolist())

        # Append the similarity scores and texts to the lists
        for k in range(len(df2)):
            text_df1.append(batch1.iloc[j]['Text'])
            text_df2.append(df2.iloc[k]['Text'])
            similarities.append(similarity_row[0][k])

        # Calculate and print the elapsed time for this row
        row_elapsed_time = time.time() - row_start_time
        print(f"Time elapsed for row: {row_elapsed_time:.2f} seconds")

        # Calculate and print the estimated remaining time for the rest of the rows in this batch
        completed_rows = j + 1
        total_rows = len(batch1)
        average_time_per_row = row_elapsed_time / completed_rows
        estimated_remaining_time_rows = average_time_per_row * (total_rows - completed_rows)
        print(f"Estimated remaining time for the rest of the rows in this batch: {estimated_remaining_time_rows:.2f} seconds")

    # Clear batch1 from memory
    del batch1
    gc.collect()

    # Calculate and print the elapsed time for this batch
    batch_elapsed_time = time.time() - batch_start_time
    print(f"Time elapsed for batch: {batch_elapsed_time:.2f} seconds")

    # Calculate and print the estimated remaining time for the rest of the batches
    completed_batches = i + 1
    total_batches = 3
    average_time_per_batch = (time.time() - start_time) / completed_batches
    estimated_remaining_time_batches = average_time_per_batch * (total_batches - completed_batches)
    print(f"Estimated remaining time for the rest of the batches: {estimated_remaining_time_batches:.2f} seconds")

# Create the DataFrame from the lists
similarity_scores = pd.DataFrame({
    'Text_df1': text_df1,
    'Text_df2': text_df2,
    'Similarity': similarities
})

print(similarity_scores)
```
