```python
data = pd.read_excel("/content/ER PYCARET.xlsx")
data
```

```python
data['DATE'] = pd.to_datetime(data['DATE'])
data.head(12)
data['MA12'] = data['COUNT'].rolling(12).mean()
```

```python
# plot the data and MA
import plotly.express as px
fig = px.line(data, x="DATE", y=["COUNT", "MA12"], template = 'plotly_dark')
fig.show()
# extract month and year from dates**
data['Month'] = [i.month for i in data['DATE']]
data['Year'] = [i.year for i in data['DATE']]
```

```python
# create a sequence of numbers
data['Series'] = np.arange(1,len(data)+1)

# drop unnecessary columns and re-arrange
data.drop(['DATE', 'MA12'], axis=1, inplace=True)
data = data[['Series', 'Year', 'Month', 'COUNT']]
```

```python
# check the head of the dataset**
data
```

```python
# split data into train-test set
train = data[data['Series'] < 7]
test = data[data['Series'] >= 7]
```

```python
s = setup(data = train, test_data = test,
          target = 'COUNT', 
          fold_strategy = 'timeseries',
          numeric_features = ['Year', 'Series'],
          fold = 5,
          transformation= True,
          transform_target = True,
          normalize=False,
          session_id = 123, silent = True,
          pca=False,
          remove_multicollinearity=True,
          use_gpu=True,
          log_experiment = True, experiment_name = 'COUNT')
```
